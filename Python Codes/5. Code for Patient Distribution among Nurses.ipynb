# Importing Libraries and Reading Model Data

```python
#Import Necessary Libraries
import pandas as pd
import pickle
import string
import random
```

```python
#Reading Model Data
model_data = pd.read_csv("ModelCovidData.csv")
#Dropping Unnecessary Column that was automatically generated when saving
model_data.drop("Unnamed: 0", axis = 1, inplace = True)
```

```python
#Displaying Data
model_data
```

```python
#Separating Input (X) and Output (y) values
X = model_data[["Age", "Sex", "Smoking", "Pregnancy", "Pneumonia", "Asthma", "Diabetes", "Obesity", "Hypertension", "Cardiovascular", "CODP", "Renal_Chronic", "Immunosuppression", "Other_Diseases"]].values
y = model_data.values[:, 14]
```

# Nurse Generation

```python
#Creating Nurse Dictionary
nurse_dict = {}
#Getting Input from User for the Number of Nurses
num_nurses = int(input("Enter Number of Nurses: "))
#Creating a String containing Letters which represents Names of Nurses
nurse_names = string.ascii_uppercase
#Creating Nurse Lists. If we have more nurses than the number of letters, we use second nurse list to distinguish them
nurse_list1 = list(nurse_names)
nurse_list2 = list(nurse_names)
#Randomly Shuffling second Nurse List in order to prevent Repititive Letters in the Nurse Names
random.shuffle(nurse_list2)
j = 0
#Creating Nurse Dictionary using for loop and Setting empty lists for each nurse keys. These lists are for storing patients
for i in range(num_nurses + int(num_nurses/len(nurse_names))):
    #Checking whether the iteration number passed number of nurses or not
    if j < len(nurse_list1):
        nurse_dict[nurse_list1[j]] = []
        j = j + 1
    #If iteration number passed number of nurses, we concatenate first and second nurse list values to distinguish nurse names
    else:
        j = 0
        for k in range(len(nurse_list1)):
            nurse_list1[k] = nurse_list1[k] + str(nurse_list2[k])
```

```python
#Displaying Nurse List
nurse_dict
```

# Patient Distribution

```python
#Opening the Model File
file = open('ModelCovid.txt', 'rb')
#Loading the Model to the variable called model
model = pickle.load(file)
#Closing the file
file.close()
```

```python
#Making Nurse Names as a List so that we can access to Nurse Dictionary Keys which are Nurse Names
nurse_names_list = list(nurse_dict)
#Getting Input for Number of Patients from User
num_patients = int(input("Enter Number of Patients: "))

if num_patients > (3.5*num_nurses):
    print("Number of Patients Exceeds Number of Nurses!\nDecrease Number of Patients!")
else:
    #Making Criticality Levels List and Naming it as patients
    patients = list(model.predict(X[:num_patients, :]) + 1)
    #Defining the Iterators
    j = 0
    k = 0
    #Iterating over each patient values and Appending them to their Nurses based on certain conditions
    while True:
        if j <= (len(nurse_dict) - 1):
            if sum(nurse_dict[nurse_names_list[j]]) == 0:
                nurse_dict[nurse_names_list[j]].append(patients[k])
                del patients[k]
            elif sum(nurse_dict[nurse_names_list[j]]) == 1 and (patients[k] == 1 or patients[k] == 2 or patients[k] == 3):
                nurse_dict[nurse_names_list[j]].append(patients[k])
                del patients[k]
            elif sum(nurse_dict[nurse_names_list[j]]) == 2 and (patients[k] == 1 or patients[k] == 2 or patients[k] == 3):
                nurse_dict[nurse_names_list[j]].append(patients[k])
                del patients[k]
            elif sum(nurse_dict[nurse_names_list[j]]) == 3 and (patients[k] == 1 or patients[k] == 2 or patients[k] == 3):
                nurse_dict[nurse_names_list[j]].append(patients[k])
                del patients[k]
            elif sum(nurse_dict[nurse_names_list[j]]) == 4 and (patients[k] == 1 or patients[k] == 2 or patients[k] == 3):
                nurse_dict[nurse_names_list[j]].append(patients[k])
                del patients[k]
            elif sum(nurse_dict[nurse_names_list[j]]) == 5 and (patients[k] == 1 or patients[k] == 2):
                nurse_dict[nurse_names_list[j]].append(patients[k])
                del patients[k]
            elif sum(nurse_dict[nurse_names_list[j]]) == 6 and (patients[k] == 1):
                nurse_dict[nurse_names_list[j]].append(patients[k])
                del patients[k]
            if len(patients) == 0:
                print("Patients were Distributed among Nurses!")
                break
        else:
            j = 0
            if sum(nurse_dict[nurse_names_list[j]]) == 0:
                nurse_dict[nurse_names_list[j]].append(patients[k])
                del patients[k]
            elif sum(nurse_dict[nurse_names_list[j]]) == 1 and (patients[k] == 1 or patients[k] == 2 or patients[k] == 3):
                nurse_dict[nurse_names_list[j]].append(patients[k])
                del patients[k]
            elif sum(nurse_dict[nurse_names_list[j]]) == 2 and (patients[k] == 1 or patients[k] == 2 or patients[k] == 3):
                nurse_dict[nurse_names_list[j]].append(patients[k])
                del patients[k]
            elif sum(nurse_dict[nurse_names_list[j]]) == 3 and (patients[k] == 1 or patients[k] == 2 or patients[k] == 3):
                nurse_dict[nurse_names_list[j]].append(patients[k])
                del patients[k]
            elif sum(nurse_dict[nurse_names_list[j]]) == 4 and (patients[k] == 1 or patients[k] == 2 or patients[k] == 3):
                nurse_dict[nurse_names_list[j]].append(patients[k])
                del patients[k]
            elif sum(nurse_dict[nurse_names_list[j]]) == 5 and (patients[k] == 1 or patients[k] == 2):
                nurse_dict[nurse_names_list[j]].append(patients[k])
                del patients[k]
            elif sum(nurse_dict[nurse_names_list[j]]) == 6 and (patients[k] == 1):
                nurse_dict[nurse_names_list[j]].append(patients[k])
                del patients[k]
            if len(patients) == 0:
                print("Patients were Distributed among Nurses!")
                break
        j = j + 1
```
